### Key Findings

- Team mentions are relatively balanced between France and Canada, suggesting neither side dominates the sample.
- Referee-related mentions occur infrequently relative to team mentions, indicating that officiating is discussed less often than team performance in this sample.
- Mbappe receives substantially more mentions than any other player, likely due to his contributions during the match.
- Mbappe and Davies receive the most player mentions for their respective teams, indicating that player identity may influence the volume of fan discussion.
- Goal-related tweets account for the largest share of event-tagged tweets, suggesting goals may be a primary driver of fan engagement.
- Non-event tweets ("none") remain common, indicating that fan discussion also centers on player performances and general match flow.

### Hypotheses for Sentiment Analysis

- Goal-related events may generate stronger sentiment than other event types.
- High-profile players may receive more extreme sentiment than less-discussed players.
- Sentiment may vary over time in response to major match events.
